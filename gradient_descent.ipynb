```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def f(x):
    return x**3 - 3*x

def df(x):
    return 3*x**2 - 3
```

```python
def gradient_descent(starting_point, learning_rate, epochs):
    x = starting_point
    print(f"Initial x: {x:.4f}, f(x) = {f(x):.4f}")
    for i in range(epochs):
        print(f"epoch {i+1}: x = {x:.4f}, f(x) = {f(x):.4f}, df(x) = {df(x):.4f}, df(x)*learning_rate = {df(x)*learning_rate:.4f}")
        x = x - df(x)*learning_rate
    return x
```

```python
learning_rate = 0.01
epochs = 100
starting_point = 15
minimum = gradient_descent(starting_point, learning_rate, epochs)
print(f"Local minimum occurs at x = {minimum:.4f}, f(x) = {f(minimum):.4f}")
```

```python
x_vals = np.linspace(-5,5,100)
y_vals = f(x_vals)
plt.plot(x_vals,y_vals,label="f(x) = x^3 - 3x")
plt.scatter(minimum,f(minimum),color='red',label='Local Minima')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent Visualization")
plt.legend()
plt.show()
```
